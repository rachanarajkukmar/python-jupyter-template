```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate y as the square of x
y = x**2

# Create the plot
plt.plot(x, y)

# Customize the plot with a title and axis labels
plt.title("Square of X Values")         # Title of the plot
plt.xlabel("X Values")                  # Label for the x-axis
plt.ylabel("Y Values (X Squared)")      # Label for the y-axis

# Add a grid to the plot
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Create the bar plot with different colors
plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])

# Add a title
plt.title("Bar Plot with Different Colors")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y)

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Add a title to the plot
plt.title("Scatter Plot with Annotations")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title("Programming Language Popularity")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the charts
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X Values")
axs[0, 0].set_ylabel("Y Values (X Squared)")

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Third subplot: Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y)
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f"({scatter_x[i]}, {scatter_y[i]})", (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(0, 10), ha='center')
axs[1, 0].set_title("Scatter Plot")

# Fourth subplot: Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
